```python
from sklearn.preprocessing import StandardScaler

scale = StandardScaler()

dados_vendas_censo[['Metragem', 'Quartos', 'Banheiros', 'latitude', 'longitude', 'V005', 'V007', 'V009', 'V011']] = scale.fit_transform(
    dados_vendas_censo[['Metragem', 'Quartos', 'Banheiros', 'latitude', 'longitude', 'V005', 'V007', 'V009', 'V011']])

dados_vendas_censo[['Metragem', 'Quartos', 'Banheiros',
                    'latitude', 'longitude', 'V005', 'V007', 'V009', 'V011']]
```

```python
X = dados_vendas_censo[['Metragem', 'Quartos', 'Banheiros',
                        'latitude', 'longitude', 'V005', 'V007', 'V009', 'V011']]
Y = dados_vendas_censo['Valor_anuncio']

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.25, random_state=54)
```

```python
from sklearn.neural_network import MLPRegressor

mlpr = MLPRegressor()

mlpr.fit(X_train, Y_train)

Y_train_pred = mlpr.predict(X_test)

mae = mean_absolute_error(Y_test, Y_train_pred)
r2 = r2_score(Y_test, Y_train_pred)
(mae, r2)
```

```python
from sklearn.svm import SVR

svr = SVR()

svr.fit(X_train, Y_train)

Y_train_pred = svr.predict(X_test)

mae = mean_absolute_error(Y_test, Y_train_pred)
r2 = r2_score(Y_test, Y_train_pred)
(mae, r2)
```

```python
from sklearn.neighbors import KNeighborsRegressor

knr = KNeighborsRegressor()

knr.fit(X_train, Y_train)

Y_train_pred = knr.predict(X_test)

mae = mean_absolute_error(Y_test, Y_train_pred)
r2 = r2_score(Y_test, Y_train_pred)
(mae, r2)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=1000)

lr.fit(X_train, Y_train)

Y_train_pred = lr.predict(X_test)

mae = mean_absolute_error(Y_test, Y_train_pred)
r2 = r2_score(Y_test, Y_train_pred)
(mae, r2)
```

```python
ls = LinearRegression()
ls.fit(X_train, Y_train)

Y_train_pred = ls.predict(X_test)

mae = mean_absolute_error(Y_test, Y_train_pred)
r2 = r2_score(Y_test, Y_train_pred)
(mae, r2)
```

```python
from sklearn.model_selection import GridSearchCV

knr = KNeighborsRegressor()

params = {
    'n_neighbors': [5, 6, 7, 8, 9, 10],
    'weights': ['uniform', 'distance'],
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
}

gs = GridSearchCV(knr, params, cv=5, n_jobs=-1, scoring='r2')

gs.fit(X, Y)
print('Best params MLPR:\n' + str(gs.best_params_) + '\n')
print('Best score MLPR:\n' + str(gs.best_score_) + '\n')
```

```python
knr = KNeighborsRegressor(algorithm='ball_tree', n_neighbors=8, weights='distance')

knr.fit(X_train, Y_train)

Y_train_pred = knr.predict(X_test)

mae = mean_absolute_error(Y_test, Y_train_pred)
r2 = r2_score(Y_test, Y_train_pred)
(mae, r2)
```
