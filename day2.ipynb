### Desafio 1 & 2

```python
dados_vendas.drop(["Valor", "Moeda", "Valor_anuncio",
                  "Cidade", "Tipo_anuncio"], axis=1, inplace=True)
dados_vendas
```

```python
dados_vendas.rename(columns={"Valor_anuncio_float": "Valor"}, inplace=True)
dados_vendas
```

```python
dados_vendas['Valor_M'] = dados_vendas['Valor'].apply(lambda x: x/1000000)
dados_vendas
```

```python
sns.set_context("poster", font_scale=1, rc={"grid.linewidth": 5, "axes.linewidth": 5, })
plt.figure(figsize=(13, 8))
ax = sns.histplot(data=dados_vendas, x="Valor_M", kde=True)
ax.set_title("Quntidade X Valor dos Imóveis")
ax.set_xlabel("Valor em Milhões")
ax.set_ylabel("Quantidade de Imóveis")
ax.legend(['Densidade de Probabilidade (KDE)'], prop={'size': 20})
plt.xlim((0, 12))
```

### Desafio 3

```python
dados_vendas.groupby("Bairro").mean()
```

```python
media_valores_bairro = dados_vendas.groupby(
    'Bairro').mean()['Valor']
media_metragem_bairro = dados_vendas.groupby('Bairro').mean()['Metragem']


valor_metragem = media_valores_bairro/media_metragem_bairro
valor_metragem
```

```python
valor_metragem_sample = pd.DataFrame(valor_metragem.sample(20)).rename(
    columns={0: "Valor"})
valor_metragem_sample.sort_values('Valor', inplace=True)
valor_metragem_sample.reset_index(inplace=True)
valor_metragem_sample
```

```python
xMax = 13000
fig, ax = plt.subplots(figsize=(16, 10), dpi=300)
ax.hlines(y=valor_metragem_sample.index, xmin=0, xmax=xMax, color='gray',
          alpha=0.7, linewidth=1, linestyles='dashdot')
ax.scatter(y=valor_metragem_sample.index, x=valor_metragem_sample.Valor,
           s=75, color='firebrick', alpha=0.7)

ax.set_title('Preço por Metragem', fontdict={'size': 22})
ax.set_xlabel('Reais por metro quadrado')
ax.set_yticks(valor_metragem_sample.index)
ax.set_yticklabels(valor_metragem_sample.Bairro.str.title(), fontdict={
                   'horizontalalignment': 'right'})
ax.set_xlim(1500, xMax)
ax.margins(0.05)
plt.show()
```
